```python
#generate a 3D spline with a given number of 3D points that passes through a given set of points

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.interpolate import splprep, splev

# generate a set of points
n = 4
x = np.random.rand(n)
y = np.random.rand(n)
z = np.random.rand(n)

# generate a spline
tck, u = splprep([x, y, z], s=0)
unew = np.linspace(0, 1, 1000)
out = splev(unew, tck)

# plot the spline
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')  # Use add_subplot instead of gca
ax.plot(x, y, z, 'ro')
ax.plot(out[0], out[1], out[2], 'b')
plt.show()

```

```python
import numpy as np
from scipy.interpolate import BSpline, make_interp_spline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.interpolate import BSpline
from mpl_toolkits.mplot3d import Axes3D

# Crea dei punti di controllo
control_points = np.array([[0, 0, 0], [0, 1, 0], [1, 1, 1], [1, 0, 1]])

# Crea una curva spline 3D
k = 2  # grado della spline
t = np.linspace(0, 1, control_points.shape[0], endpoint=True)
t = np.append(np.zeros(k), t)  # aggiunge zeri all'inizio
t = np.append(t, np.ones(k))  # aggiunge uni alla fine
print(t.size)
print(control_points.shape[0])
spline = make_interp_spline(t, control_points, k=k)
tnew = np.linspace(0, 1, 100)
control_points_new = spline(tnew)
print(spline, len(control_points_new))

# Visualizza i punti di controllo e la curva spline
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(control_points[:, 0], control_points[:, 1], control_points[:, 2], 'ro', ms=5)
ax.plot(control_points_new[0, :], control_points_new[1, :], control_points_new[2, :], 'b-', lw=3)
plt.show()
```

```python
import numpy as np
from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt
x = np.arange(10)
y = np.sin(x)
cs = CubicSpline(x, y)
xs = np.arange(-0.5, 9.6, 0.1)
fig, ax = plt.subplots(figsize=(6.5, 4))
ax.plot(x, y, 'o', label='data')
ax.plot(xs, np.sin(xs), label='true')
ax.plot(xs, cs(xs), label="S")
ax.plot(xs, cs(xs, 1), label="S'")
ax.plot(xs, cs(xs, 2), label="S''")
ax.plot(xs, cs(xs, 3), label="S'''")
ax.set_xlim(-0.5, 9.5)
ax.legend(loc='lower left', ncol=2)
plt.show()
```
